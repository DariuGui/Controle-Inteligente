##Exemplo simples : Modelo Matemático do Neurônio.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
u = np.linspace(-1, 1, N)
ruido = np.random.normal(0, 1, N)
y = (2*u + 3) + 0.3*ruido

plt.figure(figsize = (7, 5))
plt.scatter(u, y)
plt.show()
```

```python
def f(x,y):
    return x**2 + y**2

v1 = np.linspace(-2, 2, 20); v2 = np.linspace(-2, 2, 20)
U, V = np.meshgrid(v1, v2)

Z = f(U, V)
fig = plt.figure(figsize=(10, 7))
ax =plt.axes(projection = '3d')
ax.plot_surface(U, V, Z, alpha = 0.4)
plt.show()
```

```python
a = 0
b = 0
epc = 20
eta = 0.5

for k in range(epc):
  y_pred = a*u + b
  e = (y - y_pred)
  Da = -(2/N)*sum(e*u)
  Db = -(2/N)*sum(e)
  a = a - eta*Da
  b = b - eta*Db

print('O modelo ajustado é: y = %.4f u + %.4f'%(a,b))

y_pred = a*u + b
plt.scatter(u,y)
plt.plot(u,y_pred,'r')
plt.show()
```

```python
A = np.array([[sum(u*u), sum(u)],[sum(u), N]])
b = np.array([sum(y*u),sum(y)])
v = np.linalg.solve(A,b)

print('O modelo ajustado é: y = %.4f u + %.4f'%(v[0],v[1]))
y_prd = v[0]*u + v[1]
plt.scatter(u,y)
plt.plot(u,y_prd,'--k')
plt.show()
```
